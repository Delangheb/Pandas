```python
import pandas as pd
import os
```

```python
df = pd.read_pickle(os.path.join('demos', 'data_frame.pickle'))
```

```python
#Simplest plot
acquisition_years = df.groupby('acquisitionYear').size()
acquisition_years.plot()
```

```python
#Leverage Matplotlib
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams.update({'figure.autolayout' : True,
                 'axes.titlepad' : 20})

fig = plt.figure()
# Figure with 1 row, 1 column and we're adding the fist figure
subplot = fig.add_subplot(1, 1, 1)
acquisition_years.plot(ax=subplot)
fig.show()
```

```python
fig = plt.figure()
subplot = fig.add_subplot(1, 1, 1)
acquisition_years.plot(ax=subplot)
subplot.set_xlabel("Acquisition Years")
subplot.set_ylabel("Artworks Acquired")
fig.show()
```

```python
fig = plt.figure()
subplot = fig.add_subplot(1, 1, 1)
acquisition_years.plot(ax=subplot)
subplot.set_xlabel("Acquisition Years")
subplot.set_ylabel("Artworks Acquired")
subplot.locator_params(nbins=40, axis='x')
fig.show()
```

```python
fig = plt.figure()
subplot = fig.add_subplot(1, 1, 1)
acquisition_years.plot(ax=subplot, rot= 45)
subplot.set_xlabel("Acquisition Years")
subplot.set_ylabel("Artworks Acquired")
subplot.locator_params(nbins=40, axis='x')
fig.show()
```

```python
#Y-Axis in Log Scale
fig = plt.figure()
subplot = fig.add_subplot(1, 1, 1)
acquisition_years.plot(ax=subplot, rot= 45, logy=True)
subplot.set_xlabel("Acquisition Years")
subplot.set_ylabel("Artworks Acquired")
subplot.locator_params(nbins=40, axis='x')
fig.show()
```

```python
#Add Grid
fig = plt.figure()
subplot = fig.add_subplot(1, 1, 1)
acquisition_years.plot(ax=subplot, rot= 45, logy=True, grid = True)
subplot.set_xlabel("Acquisition Years")
subplot.set_ylabel("Artworks Acquired")
subplot.locator_params(nbins=40, axis='x')
fig.show()
```

```python
#Set Fonts
title_font = { 'family' : 'source sans pro',
               'color' : 'darkblue',
               'weight' : 'normal',
               'size' : 20}

labels_font = { 'family' : 'consolas',
               'color' : 'darkred',
               'weight' : 'normal',
               'size' : 16}
```

```python
fig = plt.figure()
subplot = fig.add_subplot(1, 1, 1)
acquisition_years.plot(ax=subplot, rot= 45, logy=True, grid = True)
subplot.set_xlabel("Acquisition Years", fontdict=labels_font, labelpad = 10)
subplot.set_ylabel("Artworks Acquired", fontdict=labels_font)
subplot.locator_params(nbins=40, axis='x')
subplot.set_title('Tate Galery Acquisitions', fontdict=title_font)
fig.show()
```

```python
#Save plot to a file
fig.savefig(os.path.join('demos','plot.png'))
fig.savefig(os.path.join('demos','plot.svg'), format = 'svg') #Vector graphic
```
